```python
# Demonstrates a while loop, counting down
i = 3
while i != 0:
    print("meow")
    i = i -1
```

```python
# Demonstrates a while loop, counting up from 1
i = 1
while i <= 3:
    print("meow")
    i = i + 1
# Demonstrates (more succint) incrementation
# i+=1
```

```python
# Demonstrates a for loop, using a list
for i in [0, 1, 2]:
    print("meow")
```

```python
# Demonstrates a for loop, using range
for i in range(3):
    print("meow")

# Demonstrates a for loop, with _ as variable
for _ in range(3):
    print("meow")
```

```python
# Demonstrates str multiplication
print("meow\n" * 3, end="")
```

```python
# Introduces continue, break
while True:
    n = int(input("what's n? "))
    if n <= 0:
        continue
    else:
        break

for _ in range(n):
    print("meow")
```

```python
# Demonstrates defining functions

def main():
    meow(get_n())

def get_n():
    while True:
        n = int(input("n: "))
        if n > 1:
            return n

def meow(n):
    for _ in range(n):
        print("meow")

main()
```

```python
# Demonstrates iterating over a list

students = ["Hermione", "Harry", "Ron"]

for student in students:
    print(student)

# Demonstrates iterating over and indexing into a list

students = ["Hermione", "Harry", "Ron"]

for i in range(len(students)):
    print(i+1, students[i])
```

```python
# Demonstrates indexing into a dict

students = {
    "Hermione": "Gryffindor",
    "Harry": "Gryffindor",
    "Ron": "Gryffindor",
    "Draco": "Slytherin"
}

for student in students:
    print(student, students[student], sep=", ") #playing around with print() arg
```

```python
# Demonstrates iterating over a list of dict objects

students = [
    {"name": "Hermione", "house": "Gryffindor", "patronus": "Otter"},
    {"name": "Harry", "house": "Gryffindor", "patronus": "Stag"},
    {"name": "Ron", "house": "Gryffindor", "patronus": "Jack Russell terrier"},
    {"name": "Draco", "house": "Slytherin", "patronus": None} #Using None
]

for student in students:
    print(student["name"], student["house"], student["patronus"], sep=", ")

# print(students[0]["name"]) #[Index][Key]
```

```python
# Prints square of bricks using a function with nested loops

def main():
    print_square(get_num())

def get_num():
    n = int(input("Enter n: "))
    return n

def print_square(size):
    for i in range(size):
        for j in range(size):
            print("#", end="")
        print()
main()
```
