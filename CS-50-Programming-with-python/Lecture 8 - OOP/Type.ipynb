```python
# Prints the type of an integer
print(type(50))

# Prints the type of a string
print(type("hello, world"))

# Prints the type of a list
print(type([]))

# Prints the type of a list
print(type(list()))

# Prints the type of a dictionary
print(type({}))

# Prints the type of a dictionary
print(type(dict()))
```
