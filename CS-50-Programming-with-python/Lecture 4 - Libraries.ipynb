```python
# Demonstrates import and random.choice

import random

coin = random.choice(["heads", "tails"])
print(coin)
```

```python
# Demonstrates from

from random import choice

coin = choice(["heads", "tails"])
print(coin)
```

```python
# Demonstrates randint

import random

numbers = [random.randint(6,7),random.randint(6,7)]
for number in numbers:
    print(number)
```

```python
# Demonstrates shuffle

import random

cards = ["jack","queen","king"]
random.shuffle(cards)
for card in cards:
    print(card)
```

```python
# Demonstrates statistics

import statistics

print(statistics.mean([100,90]))
```

```python
# Demonstrates sys.argv

import sys

print("hello, my name is", sys.argv[0])
```

```python
# Demonstrates IndexError

import sys

try:
    print("hello, my name is", sys.argv[1])
except IndexError:
    print("too few arguments")
```

```python
# Demonstrates iterating over JSON

import json
import sys
import requests

response = requests.get("https://itunes.apple.com/search?entity=song&limit=10&term=theweeknd")
o = response.json()
for result in o["results"]:
    print(result["trackName"])
```
