```python
import re
```

```python
# Validates email address by checking for @ with regex

email = input("What's your email? ").strip()

if re.search("@", email):
    print("valid")
else:
    print("Invalid")
```

```python
# Adds .*

email = input("What's your email? ").strip()

if re.search(".*@.*", email):
    print("Valid")
else:
    print("Invalid")
```

```python
# Changes * to +

email = input("What's your email? ").strip()

if re.search(".+@.+", email):
    print("Valid")
else:
    print("Invalid")
```

```python
# Adds \.edu

email = input("What's your email? ").strip()

if re.search(r".+@.+\.edu", email):
    print("Valid")
else:
    print("Invalid")
```

```python
# Adds ^ and $ to regex

email = input("What's your email? ").strip()

if re.search(r"^.+@.+\.edu$", email):
  print("Valid")
else:
  print("Invalid")
```

```python
# Adds character class

email = input("What's your email? ").strip()

if re.search(r"^[a-zA-Z0-9_]+@[a-zA-Z0-9_]+\.edu$", email):
  print("Valid")
else:
  print("Invalid")
```

```python
# Replaces character class with \w

email = input("What's your email? ").strip()

if re.search(r"^\w+@\w+\.edu$", email):
  print("Valid")
else:
  print("Invalid")
```

```python
# Adds re.INGORECASE

email=input("What's your email? ").strip()

if re.search(r"^\w+@\w+\.edu$", email, re.IGNORECASE):
  print("Valid")
else:
  print("Invalid")
```

```python
# Adds optional subdomain

email = input("What's your email? ").strip()

if re.search(r"^\w+@(\w+\.)?\w+\.edu$", email, re.IGNORECASE):
  print("Valid")
else:
  print("Invalid")
```

```python
# Reformats "last, first" as "first last"

name = input("What's your name? ").strip()
if "," in name:
    last, first = name.split(", ")
    name = f"{first} {last}"
print(f"hello, {name}")
```

```python
# Uses re.search

name =  input("What's your name? ").strip()
matches = re.search(r"^(.+), (.+)$", name)
if matches:
    name = matches.group(2) + " " + matches.group()
print(f"hello, {name}")
```

```python
# Uses .group

name = input("What's your name? ").strip()
matches =  re.search(r"^(.+), (.+)$", name)
if matches:
    name = matches.group(2) + " " + matches.group(1)
print(f"hello, {name}")
```

```python
# Uses walrus operator

name =  input("What's your name? ").strip()
if matches := re.search(r"^(.+), (.+)$", name):
    name = matches.group(2) + " " + matches.group(1)
print(f"hello, {name}")
```

```python
# Extracts Twitter username from URL using str.replace

url = input("URL: ").strip()

username = url.replace("https://twitter.com/", "")
print(f"Username: {username}")
```

```python
# Uses re.sub

url = input("URL: ").strip()

username = re.sub(r"^https://twitter\.com/", "", url)

print(f"Username: {username}")
```

```python
# Allows for http, no protocol, and www.

url = input("URL: ").strip()

username = re.sub(r"^(https?://)?(www\.)?twitter\.com/", "", url)
print(f"Username: {username}")
```

```python
# Uses capture group

import re

url = input("URL: ").strip()

matches = re.search(r"^(https?://)?(?:www\.)?twitter\.com/(.+)$", url, re.IGNORECASE)
if matches:
    print("Username:", matches.group(2))
```
