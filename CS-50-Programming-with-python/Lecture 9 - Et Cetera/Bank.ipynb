```python
# Demonstrates UnboundLocalError

balance = 0

def main():
    print("Balance: ", balance)
    deposit(100)
    withdraw(50)
    print("Balance: ", balance)

def deposit(n):
    balance += n

def withdraw(n):
    balance -= n

if __name__ == "__main__":
    main()
```

```python
# Uses global

balance = 0

def main():
    print("Starting Balance: ", balance)
    deposit(100)
    withdraw(50)
    print("Updated Balance: ", balance)

def deposit(n):
    global balance
    balance += n

def withdraw(n):
    global balance
    balance -= n

if __name__ == "__main__":
    main()
```
