```python
# Prints arbitrarily many args in uppercase

def main():
    yell("This", "is", "CS50")

def yell(*words):
    uppercased = []
    for word in words:
        uppercased.append(word.upper())
    print(*uppercased)

if __name__ == "__main__":
    main()
```

```python
# Uses map

def main():
    yell("This", "is", "CS50")

def yell(*words):
    uppercased = map(str.upper, words)

    # Uses list comprehension
    # uppercased = [arg.upper() for arg in words]
    
    print(*uppercased)

if __name__ == "__main__":
    main()
```
